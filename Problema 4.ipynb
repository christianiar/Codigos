```python
def make(a):
    s = 0
    l = len(a)
    b = 0
    real = 0
    while b<l:
        real = real*10
        real = real+a[b]
        b+=1
    return real
```

```python
n_max = 999*999
n_min = 100*100
poli = []
while n_min<n_max:
    c = []
    c = [int(x) for x in str(n_min)]
    if c[0]==c[-1] and c[1]==c[-2] and c[2]==c[-3]:
        poli.append(make(c))
    n_min+=1
```

```python
for i in range(len(poli)):
    n1 = 100
    n2= 999
    while n1<n2:
        check = poli[-i]%n1
        if check==0:
            n3 = n1+1
            while n3<n2:
                check2 = poli[-i]%n3
                if check2==0:
                    temp = n1*n3
                    if temp==poli[-i]:
                        print(poli[-i],n1,n3)
                        
                n3+=1
        n1+=1
```
