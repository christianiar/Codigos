```
def divisores(n):
  pri = int(n/2)+1
  count = 0
  while pri>0:
    if n%pri == 0:
      count +=pri
    pri-=1
  return count
```

```
tam = 10000
num = []
for i in range(tam):
  num.append(divisores(i))
```

```
soma = 0
for i in range(len(num)):
  if num[i]<len(num):
    a = num[num[i]]
    if a == i and num[a]!=a:
      print(i,a,num[num[i]],num[i])
      soma+=a
```
